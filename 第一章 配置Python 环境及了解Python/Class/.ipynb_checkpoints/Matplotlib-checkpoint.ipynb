```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,5,0.2) # 绘制X轴 0-5 间隔0.2
y = np.sin(x)
plt.plot(x,y) #绘制图形
plt.show() # 显示图形
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,5,0.2) # 绘制X轴 0-5 间隔0.2
y = np.sin(x)
y2 = np.cos(x)#cos图像
plt.plot(x,y,label="sin") #绘制图形
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin&cos")
plt.legend()
plt.show() # 显示图形
```
